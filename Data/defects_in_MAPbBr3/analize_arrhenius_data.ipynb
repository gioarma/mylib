# Scatter plots

## Activation energy

```python
%%opts Scatter [frame_width=200, frame_height=200]
variables = ['Sample age (days)',
             'Bias (V)',
             'T range (K)',
             'Light source current (mA)',
             'Contact type']

hv.Layout([arr.hvplot.scatter(x=var, y='Ea (eV)', by='Sample name')*\
           arr.hvplot.errorbars(y='Ea (eV)', x=var,yerr1='dEa (eV)') for var in variables])\
.cols(2)
```

## Capture cross section

```python
%%opts Scatter [frame_width=200, frame_height=200, ylim=(arr['sigma (cm2)'].min()/10, arr['sigma (cm2)'].max()*10)]

hv.Layout([arr.hvplot.scatter(x=var, y='sigma (cm2)', by='Sample name', logy=True)*\
           arr.hvplot.errorbars(y='sigma (cm2)', x=var,yerr1='dsigma (cm2)') for var in variables])\
.cols(2)
```
